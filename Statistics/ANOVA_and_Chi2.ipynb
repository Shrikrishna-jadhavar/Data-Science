```python
#Import scipy.stats
from scipy import stats
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Shrikrishna-jadhavar/Data-Science-Material/main/Dataset/Iris.csv',
                index_col=0)
df
```

```python
df['Species']
```

```python
df['SepalLengthCm']
```

```python
f ,p = stats.f_oneway(df['SepalWidthCm'],df['SepalLengthCm'],df['PetalLengthCm'],
                       df['PetalWidthCm'])
```

```python
p
```

```python
if p < 0.05:
    print('Reject Null Hypothesis. At least one sample is different.')
else:
    print('Fail to reject Null Hypothesis.')
```

# Chi-Square test

```python
from scipy.stats import chi2_contingency
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Shrikrishna-jadhavar/Data-Science-Material/main/Dataset/chi2.csv",
                index_col=0)
df
```

```python
df['Athlete']
```

```python
obs = pd.crosstab(index=df['Athlete'],columns=df['Smoker'])
obs
```

```python
chi2,p,df,exp  = chi2_contingency(obs)
```

```python
chi2, df
```

```python
exp
```

```python
p
```

```python
if p < 0.05:
    print('Reject Null Hypothesis.Columns are dependent on each other.')
else:
    print('Fail to reject null hypothesis.')
```
